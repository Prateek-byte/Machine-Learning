```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
#to be able to plot inside the jupyter notebook
%matplotlib inline
```

```python
#plotting number of points
x = np.array([2,4,6,8,10,12,14,16])
y = x/2

plt.figure(figsize=(10,5))
plt.scatter(x,y,c='green')
plt.show()
```

```python
#Label the axes
plt.plot([1,2,3,4])
plt.ylabel('vertical')
plt.xlabel('horizontal')
plt.show()
```

```python
#Create Multiple plots with subplots
names = ['jhon', 'sanjay', 'ellen', 'sandra', 'nancy', 'shubham', 'philip']
scores = [10,15,20,30,40,50,100]

plt.figure(figsize=(20,10))

plt.subplot(131)  #131 means 1 row and 3 columns and 1st in the row
plt.bar(names,scores)

plt.subplot(132)   #131 means 1 row and 3 columns and 2nd in the row
plt.scatter(names, scores)

plt.subplot(133)
plt.plot(names,scores)

plt.suptitle('categorical plotting')
plt.show()
```

Seaborn || Scatterplot || Correlation || Boxplot || HeatMap

```python
#Seaborn
import seaborn as sns
```

```python
ourdata= pd.read_excel('Pokemon.xlsx')
ourdata.head()
```

```python
sns.lmplot(x='Attack', y='Defense', data=ourdata)  #lmplot() function is used to quickly plot the Linear Relationship between two variables.lm for linear regression model.
plt.show()
```

```python
sns.lmplot(x='Attack', y='Defense', data=ourdata, fit_reg=False)  #regression line will be removed.
```

```python
sns.lmplot(x='Attack', y='Defense', data=ourdata, fit_reg = False, hue='Legendary')
```

```python
ourdata['Legendary'].value_counts()
```

we set hue='stage' to color our points by the Pokemon's evolution stage.

```python
fig= plt.figure()
a1 = fig.add_axes([0,0,1,1])

x = np.arange(1,10)
a1.plot(x,np.exp(x), 'b')

a1.set_title('range of numbers')

#explicitly set x and y limits
plt.ylim(0,10000)
plt.xlim(0,10)

#explicitly set x and y labels
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=ourdata)
```

```python
#drop the unwanted columns
ourdata1 = ourdata.drop(['Total', 'Legendary'], axis=1)
ourdata1.head()
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=ourdata1)
```

```python
corr = ourdata1.select_dtypes(include='number').corr()
corr   #calculates correlation between numerical values
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr, annot=True)  #creating heatmap
```

```python
sns.pairplot(ourdata[['Defense', 'Attack', 'HP']], kind='scatter', diag_kind='kde')
plt.show()
```
